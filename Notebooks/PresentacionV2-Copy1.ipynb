<strong>Definición Auto Machine Learning</strong><br>
El Auto Machine Learning (AutoML) tiene como objetivo facilitar el proceso de creación de modelos de Aprendizaje Aautomático (AA) mediante la automatización de los pasos de uso común, como el preprocesamiento de funciones, la selección de modelos y el ajuste de hiperparámetros.

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>1.1 Breve descripcion del CI</strong></p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Que es el CI y para que sirve, estas preguntas se intentan responder en esta sección.
</p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Otra definición</strong><br>
La inferencia causal es el proceso por el cual se comprende que causó un resultado en particular
</p>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Importancia</strong><br>
Identificar los efectos causales es parte integrante de la investigación científica
</p>    
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Aporte</strong><br>
El aprendizaje automático fue pensado para encontrar correlaciones en los datos,
es decir va de los datos a los resultados.<br>
El CI intenta explicar esos resultados, es decir va de los resultados a los datos (características/features)
</p>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Desafío</strong><br>
Habitualmente se cuenta con datos de observaciones y/o experimentales pero si se desea evaluar nuevas alternativas se puede usar CI
</p>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Ambito</strong><br>
Educativo, social, tecnología, medicina, etc.
</p>

Las <b>causas y efectos</b> son fundamentales para el diseño de modelos de <b>Machine Learning</b>
</p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:center;border-radius:5px 5px;">

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Origen:</strong><br> 
<b><i>Judea Pearl</i></b> (creador) establece un proceso de cuatro pasos, basada en el <b>Modelo Causal Estructural (MCE)</b>:
<ol style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<li><b>Definir</b>: Expresar la cantidad objetivo Q como una función Q(M) que puede calcularse a partir de cualquier modelo M.</li>
<li><b>Suponer</b>: Formular supuestos causales utilizando lenguaje científico y representar su estructura en forma gráfica.</li>
<li><b>Identificar</b>: Si la cantidad objetivo es identificable (expresable en términos estimables)</li>
<li><b>Estimar</b>: La cantidad objetivo si es identificable, o aproximarla si no lo es.</li> 
</ol>
</p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El efecto causal de una variable <b>A</b> en <b>Y</b>, se define como el cambio esperado en <b>Y</b> debido a un cambio en <b>A</b><br>
Utilizando la notación <b>do-calculus</b>, el efecto causal medio puede escribirse como <b>E[Y|do(a)]</b>     
</p>
<p style="font-size:12px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:px 5px;"><i>
<a href="https://plato.stanford.edu/entries/causation-counterfactual/" target=\"_blank\">Ref Notation: The do-calculus</a></i></p>     
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
A veces, interesa el efecto causal sólo en una subpoblación o queremos comparar el efecto causal entre subpoblaciones. 
En ese caso, podemos estimar el efecto causal medio condicional (CACE) dado un conjunto de covariables  <b>X,E[Y|do(a),X]</b>     
</p>
<p style="font-size:12px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:px 5px;"><i>

<strong>Inicialización de Variables</strong></p> 

```python
np.random.seed(42)
num_users = 10000
num_months = 12
```

```python
signup_months = np.random.choice(np.arange(1, num_months), num_users) * np.random.randint(0,2, size=num_users)
```

<strong>Consulta de los meses generados</strong></p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación del dataset</strong></p> 

```python
df = pd.DataFrame({
                  'user_id'      : np.repeat(np.arange(num_users), num_months),
                  'signup_month' : np.repeat(signup_months, num_months),
                  'month'        : np.tile(np.arange(1, num_months+1), num_users), 
                  'spend'        : np.random.poisson(500, num_users*num_months)
                  })
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Los clientes que se incluyen en el grupo de tratamiento son los que se incribieron
</p>

```python
df["treatment"] = df["signup_month"] > 0
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Los clientes compran menos a final de año)
</p>

```python
df["spend"] = df["spend"] - df["month"] * 10
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Simular del efecto de tratamiento
</p>

```python
after_signup = (df["signup_month"] < df["month"]) & (df["treatment"])
df.loc[after_signup,"spend"] = df[after_signup]["spend"] + 100
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Paso 1: I. Modelado Causal del problema</strong></p> 

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Supuestos</strong></p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
La suscripción puede afectar a las transacciones posteriores, no las anteriores<br>
La actividad previa al tratamiento se considera la <b>causa</b><br>
La actividad después del tratamiento es el <b>resultado</b><br>
Esta es solamente una de las alternativas, existen otras variables que afectan a la suscripción
</p>

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Mes de suscripción</strong></p> 

```python
i=3
```

```python
causal_graph = """digraph {
                           treatment[label="Program Signup in month 3"];
                           pre_spends;
                           post_spends;
                           Z->treatment;
                           pre_spends -> treatment;
                           treatment->post_spends;
                           signup_month->post_spends;
                           signup_month->treatment;
               }"""
```

```python
df_i_signupmonth = (
    df[df.signup_month.isin([0, i])]
    .groupby(["user_id", "signup_month", "treatment"])
    .apply(
        lambda x: pd.Series(
            {
                "pre_spends": x.loc[x.month < i, "spend"].mean(),
                "post_spends": x.loc[x.month > i, "spend"].mean(),
            }
        )
    )
    .reset_index()
)
```

```python
dfTmp = pd.DataFrame(df_i_signupmonth)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Visualizar Dataframe</strong></p> 

```python
dfTmp.loc[dfTmp['signup_month'] == 0].sample(3)
```

```python
dfTmp.loc[dfTmp['signup_month'] != 0].sample(3)
```

```python
del dfTmp 
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Creación & Visualición del Modelo</strong></p> 

```python
model = dowhy.CausalModel(data=df_i_signupmonth,
                          graph=causal_graph.replace("\n", " "),
                          treatment="treatment",
                          outcome="post_spends")
model.view_model()
```

![Alt Text](causal_model.png)

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Paso 2: Identificar el objetivo del modelo</strong></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<b>proceed_when_unidentifiable = True</b><br>
Indicador binario, define que debe realizarse en presencia de (potenciales) factores de confusión no observados<br><br>
<b><i>Confusión (confounding)</i></b><br>
La <b>confusión</b> es una <b>superposición de efectos</b>.<br>
Los efectos estudiados sobre un resultado determinado se mezclan con los efectos de otro factor o varios, lo que provoca la distorsión de la verdadera relación.</p>
<p style="font-size:13px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:5px 5px;">
<a href="https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3503514/#:~:text=What%20is%20confounding%3F,distortion%20of%20the%20true%20relationship" target=\"_blank\">Assessing bias: the importance of considering confounding</a>
</p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
La confusión es un concepto estrictamente causal</p>
<p style="font-size:13px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:5px 5px;">
<a https://arxiv.org/abs/2002.11631" target=\"_blank\">Zhao, Y., & Liu, Q. (2023). Causal ML: Python package for causal inference machine learning. SoftwareX, 21, 101294</a>
</p> 


<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:center;border-radius:5px 5px;">
<strong>_____________________________________________________________</strong></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<b>Average Treatment Effect</b> (ATE): medida utilizada para comparar tratamientos en experimentos aleatorios<br>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El efecto medio sobre el gasto total de los clientes que se inscribieron en el programa de fidelización en el mes i=3, en comparación con el caso en que no se hubieran inscrito </p>

```python
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Basándose en el gráfico Paso 1, DoWhy determina que es necesario condicionar <b>signup_month<b/> y <b>pre_spend</b></p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:center;border-radius:5px 5px;">
<strong>_____________________________________________________________</strong></p>

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<b>Importante:</b> No censura
<ol style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<li><b>Censura a la izquierda</b>: Si un cliente se da de alta en primer mes, no hay suficiente historia de transacciones para compararlo con clientes similares que no se dieron de alta (y así aplicar la estimación backdoor)</i>    
<li><b>Censura a la derecha</b>: Si un cliente se registra el último mes, no se dispone de suficientes transacciones futuras</i> 
</p>    

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Paso 3: Estimar el efecto causal</strong></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Estimar el efecto fijando las unidades objetivo en <b>Average Treatment Effect of the Treated</b> (att)</p>

```python
estimate = model.estimate_effect(identified_estimand,
                                 method_name='backdoor.propensity_score_matching',
                                 target_units='att')
print(estimate)
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El análisis indica el efecto medio sobre el gasto total de los clientes que se inscribieron en el programa de fidelización en el mes i=3 en comparación con los que no se han inscrito</p>

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Paso 4: Refutar la estimación</strong></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El refutador sustituye el tratamiento por una variable aleatoria independiente</p>

```python
refutation = model.refute_estimate(identified_estimand, 
                                   estimate, 
                                   method_name='placebo_treatment_refuter',
                                   placebo_type='permute', 
                                   num_simulations=20)
print(refutation)
```

![Alt Text](./img/10.png)

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Caso Práctico</strong></p>   

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Se cuenta con tres características las cuales se modelan para detectar una potencial correlación</p>
<p style="font-size:13px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:px 5px;"><i>
Fuente:    
<a href="https://www.packtpub.com/product/causal-inference-and-discovery-in-python/9781804612989" target=\"_blank\">Causal Inference and Discovery in Python</a></i></p>

```python
from itertools import combinations
import numpy as np
from scipy import stats
import networkx as nx
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
COLORS = ['#00B0F0','#FF0000']
```

```python
np.random.seed(45)

b = np.random.rand(100)

a = b + .1 * np.random.rand(100)
c = b + .3 * np.random.rand(100)
```

```python
coef, p_val = stats.pearsonr(a, c)
print(coef)
```

```python
variables = {
    'a': a,
    'b': b,
    'c': c
}

plt.figure(figsize=(12, 7))

for i, (var_1, var_2) in enumerate([('b', 'a'), ('b', 'c'), ('c', 'a')]):
    
    color = COLORS[1]
    
    if 'b' in [var_1, var_2]:
        color = COLORS[0]
    
    plt.subplot(2, 2, i + 1)
    plt.scatter(variables[var_1], variables[var_2], alpha=.8, color=color)
    
    plt.xlabel(f'${var_1}$', fontsize=16)
    plt.ylabel(f'${var_2}$', fontsize=16)

plt.suptitle('Pairwise relationships between $a$, $b$ and $c$')
plt.subplots_adjust(hspace=.25, wspace=.25)
plt.show()
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Se observa de manera clara la correlación entre las 3 variables</p>

<strong>Punto de Encuentro 02: Generación de Datos Sintéticos</strong><br> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
La falta de datos para el modelado de <b>Machine Learning</b> es un freno al proyecto el unso de <b>Inferencia Causal</b> puede suplir este punto, convinandolas con librerías estandares de generación de datos siténticos 
</p>  

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Carga de librerías de Inferencia Causal y Grafos</strong></p> 

```python
import networkx as nx
import dowhy.gcm as gcm
import numpy as np, pandas as pd
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación de Datos Originales</strong></p> 

```python
X = np.random.normal(loc=0, scale=1, size=1000)
Y = 2 * X + np.random.normal(loc=0, scale=1, size=1000)
Z = 3 * Y + np.random.normal(loc=0, scale=1, size=1000)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Visualizar Dataframe</strong></p> 

```python
data = pd.DataFrame(data=dict(X=X, Y=Y, Z=Z))
data.head()
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación del Modelo</strong></p> 

```python
causal_model = gcm.StructuralCausalModel(nx.DiGraph([('X', 'Y'), ('Y', 'Z')]))
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Agregado de ruido</strong></p> 

```python
gcm.auto.assign_causal_mechanisms(causal_model, data)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Entrenamiento del Modelo</strong></p> 

```python
gcm.fit(causal_model, data)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación de datos</strong></p> 

```python
generated_data = gcm.draw_samples(causal_model, num_samples=1000)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Visualizar Dataframe</strong></p> 

```python
generated_data.head()
```

```python
generated_data.shape
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Verificación del Modelo</strong></p> 

```python
gcm.divergence.auto_estimate_kl_divergence(data.to_numpy(), generated_data.to_numpy())
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Si las hipótesis de modelización son correctas, los datos generados tienen una distribución similar a los datos originales, la divergencia esperada debe ser mínima.</p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:center;border-radius:5px 5px;">
<strong>_____________________________________________________________</strong></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
A continuación se repite el ejecicio pero con le uso de una librería de datos sintéticos</p> 

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Carga de librerías de generación de datos sintéticos</strong></p> 

```python
from sdv.metadata import SingleTableMetadata
```

```python
from sdv.single_table import GaussianCopulaSynthesizer
```

```python
metadata = SingleTableMetadata()
```

```python
metadata.detect_from_dataframe(data=data)
```

```python
synthesizer = GaussianCopulaSynthesizer(metadata)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Entrenamiento del Modelo</strong></p> 

```python
synthesizer.fit(data)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación de datos</strong></p> 

```python
synthetic_data = synthesizer.sample(num_rows=50)
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Visualizar Dataframe</strong></p> 

```python
synthetic_data.head()
```

<p style="font-size:17px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Verificación del Modelo</strong></p> 

```python
gcm.divergence.auto_estimate_kl_divergence(data.to_numpy(), synthetic_data.to_numpy())
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
La divergencia es mayor que el modelo anterior, esto no invalida este úlitmo, son distintas formas de aproximar la solución</p>
